```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

## Nacimientos por dia 2019
https://data.buenosaires.gob.ar/dataset/nacimientos

```python
df = pd.read_csv("nacimientos.csv")
df
```

```python
df['fecha_nacimiento'] = pd.to_datetime(df['fecha_nacimiento']).dt.date
df["Count"] = 1
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.groupby("fecha_nacimiento").count().plot(y='Count', rot=90)
ax.set_title("Nacimientos por dia 2019", fontsize=16)
ax.set_ylabel("Nacimientos", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

## Matrimonios por dia 2019
https://data.buenosaires.gob.ar/dataset/matrimonios

https://cdn.buenosaires.gob.ar/datosabiertos/datasets/direccion-general-de-estadisticas-y-censos/nacimientos/nacimientos_2019.csv

```python
df = pd.read_csv("matrimonios.csv")
df
```

```python
df['fecha_matrimonio'] = pd.to_datetime(df['fecha_matrimonio']).dt.date
df["Count"] = 1
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.groupby("fecha_matrimonio").count().plot(y='Count', rot=90)
ax.set_title("Matrimonios por dia 2019", fontsize=16)
ax.set_ylabel("Matrimonios", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```
