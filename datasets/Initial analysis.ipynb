```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

## Nacimientos por dia 2019
https://data.buenosaires.gob.ar/dataset/nacimientos

```python
df = pd.read_csv("nacimientos.csv")
df
```

```python
df['fecha_nacimiento'] = pd.to_datetime(df['fecha_nacimiento']).dt.date
df["Count"] = 1
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.groupby("fecha_nacimiento").count().plot(y='Count', rot=90)
ax.set_title("Nacimientos por dia 2019", fontsize=16)
ax.set_ylabel("Nacimientos", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

# Marriage

https://data.world/guilfordcounty/15f377df-8a43-4c06-b3dc-bcffd704b528

```python
df = pd.read_csv("both_sexes.csv")
df
```

```python
df['RecDate'] = pd.to_datetime(df['RecDate']).dt.date
df["Count"] = 1
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.groupby("RecDate").count().plot(y='Count', rot=90)
ax.set_title("Matrimonios por dia", fontsize=16)
ax.set_ylabel("Matrimonios", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

```python
import datetime as dt
df = df[(df['RecDate'] > dt.date(year=2014,month=1,day=1)) & (df['RecDate'] < dt.date(year=2017,month=1,day=1))]
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.groupby("RecDate").count().plot(y='Count', rot=90)
ax.set_title("Matrimonios por dia", fontsize=16)
ax.set_ylabel("Matrimonios", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```
