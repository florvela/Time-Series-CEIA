## 1) Generar una serie de tiempo con valores aleatorios y graficarla

```python
df = pd.read_csv("datasets/TECO2.2010.2021.csv", index_col=0).sort_values(by="fechaHora")
df.fechaHora = pd.to_datetime(df["fechaHora"])
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio',rot=90)
ax.xaxis.set_major_locator(mdates.MonthLocator(1))
ax.set_title("Precio de acciones de Telecom", fontsize=16)
ax.set_ylabel("Precio (quizs est en dlares)", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

```python
df = pd.read_csv("datasets/catfish.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(by="Date")
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(x='Date', y='Total',rot=90)
ax.set_title("Venta de Catfish", fontsize=16)
ax.set_ylabel("Ventas (en 1000 pounds)", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

"Ah que bonito dataset!" me dije a mi misma cuando vi este grfico. 

Podemos ver que desde el 1986 hasta el 2004, aumentaron las ventas de catfish en general, pero se puede ver un claro ciclo, como estacional. Que segn la poca del ao, hay ms o menos ventas. Siguiendo con ese ciclo, podemos ver que las ventas empiezan a disminuir despus del 2004.

```python
df = pd.read_csv("datasets/YPFD.2000.2021.csv", index_col=0)
df["fechaHora"] = pd.to_datetime(df["fechaHora"])
df = df.sort_values(by="fechaHora")
df
```

```python
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio',rot=90)
ax.set_title("Precio de acciones de YPF", fontsize=16)
ax.set_ylabel("Precio (quizs est en dlares)", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

```python
df = df[df.ultimoPrecio > 0]
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio',rot=90)
ax.set_title("Precio de acciones de YPF", fontsize=16)
ax.set_ylabel("Precio (quizs est en dlares)", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

Los precios de acciones de YPF... Comienzan algo estable con pequeos crecimientos hasta mediados del 2018, donde hay un claro pico, que luego baja y crece; con grandes cadas en el medio... La cada ms prominente del precio se puede ver el ao 2020, que asumo que por covid: al haber menos personas saliendo de casa, hay menos uso de transportes y por ende, menos uso de gasolina. De igual manera, el precio parece que se recupera con movimientos pocos predecibles que suben y bajan.
