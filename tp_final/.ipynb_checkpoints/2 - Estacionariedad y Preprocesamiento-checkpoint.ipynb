```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, plot_predict
from statsmodels.tsa.stattools import adfuller, kpss
from scipy import stats
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
period=90

inputfile="both_sexes.csv"
print('Input file is:', inputfile)
df = pd.read_csv(inputfile)
df = df.set_index(pd.DatetimeIndex(df['RecDate']))
df.sort_index(ascending=True, inplace=True)
df["Count"] = 1
df = df["Count"]

df = df.groupby("RecDate").count()

matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(y='Count', rot=90)
ax.set_title("Matrimonios por dia", fontsize=16)
ax.set_ylabel("Matrimonios", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
plt.legend(loc='best')
plt.show()
```

# ESTACIONARIEDAD

```python
fig, ax = plt.subplots(1,2,figsize=(18,4))
plot_acf(df, lags=40, ax=ax[0])
plot_pacf(df, lags=40, method='ywm', ax=ax[1])
plt.show()
```

Se pueden ver ciclos... indicando que no es estacionaria, igual se confirma viendo los tests de adfuller y kpss

```python
adfuller(df)
```

El pvalor < 0.05 => rechazo H0 => la serie es estacionaria 

Aunque este test considera la serie diferenciada y Magui dijo que hay que tomarlo con pinzas 😃 así que veamos el test kpss...

```python
kpss(df)
```

El pvalor > 0.05 => no rechazo H0 => la serie es estacionaria.

*NOTA: los tests adfuller y kpss no funcionan bien si hay ciclos*

# PREPROCESAMIENTO

## Transformacion de box-cox

```python
original_data = df.reset_index().Count
fitted_data, fitted_lambda = stats.boxcox(original_data)
fig, ax = plt.subplots(1, 2)
# plotting the original data(non-normal) and
# fitted data (normal)
sns.distplot(original_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Non-Normal", color ="green", ax = ax[0])
 
sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Normal", color ="green", ax = ax[1])
 
 
# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(16)
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,4))
df.plot(ax=ax[0])

temp = df.reset_index()
temp["fitted_data"] = fitted_data
fitted_data = temp.set_index(pd.DatetimeIndex(temp['RecDate']))["fitted_data"]
pd.Series(fitted_data).plot(ax=ax[1])

ax[0].set_title("Antes de Box-cox")
ax[1].set_title("Después de Box-cox")
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,4))
plot_acf(fitted_data, ax=ax[0])
plot_pacf(fitted_data, method='ywm', ax=ax[1]);
```

## Transformación de Media Movil

```python
media_movil = pd.Series(fitted_data).rolling(30).mean()
media_movil.dropna(inplace=True)
media_movil.plot()
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,4))
plot_acf(media_movil, ax=ax[0])
plot_pacf(media_movil, method='ywm', ax=ax[1]);
```
