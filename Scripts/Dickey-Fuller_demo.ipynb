```python
from statsmodels.tsa.stattools import adfuller
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = norm.rvs(size=100,loc=0, scale=0.8)
y = np.cumsum(e) # genero random walk a partir del ruido
plt.figure(figsize=(12,4))
plt.plot(y)
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller(y)
```
