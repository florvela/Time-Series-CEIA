```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.arima_process import arma_generate_sample, arma_acf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

Generamos un modelo AR de orden 2

```python
ar_coef  = np.array([0.8,-0.2])
ar_ts = arma_generate_sample(ar=np.r_[1,-ar_coef], 
                             ma=[1], nsample =100)
plt.figure(figsize=(8,3))
plt.plot(ar_ts)
```

```python
fig, ax= plt.subplots(1,1,figsize=(8,3))
plot_acf(ar_ts, ax=ax)
plt.plot(arma_acf(ar=np.r_[1,-ar_coef], 
         ma=[1], lags=20))
plt.ylim(-.4,1.05)
```

```python
plot_pacf(ar_ts, method='ywm')
```

Genero un MA

```python
ma_coef = np.array([0.9,0.5,-.3])
ma_ts = arma_generate_sample(
    ar=[1], ma=np.r_[1,ma_coef], nsample =100)
plt.figure(figsize=(8,3))
plt.plot(ma_ts)
```

```python
fig, ax= plt.subplots(1,1,figsize=(8,3))
plot_acf(ma_ts, ax= ax)
plt.plot(arma_acf(ar=[1], 
         ma=np.r_[1,ma_coef], lags=20))
plt.ylim(-.4,1.05)
```

Genero un ARMA

```python
arma_ts = arma_generate_sample(
    ar=np.r_[1,-ar_coef], ma=np.r_[1,ma_coef], nsample =100)
plt.figure(figsize=(8,3))
plt.plot(arma_ts)
```

```python
fig, ax= plt.subplots(1,1,figsize=(8,3))
plot_acf(arma_ts, ax=ax)
plt.plot(arma_acf(ar=np.r_[1, -ar_coef], 
         ma=np.r_[1,ma_coef], lags=20))
plt.ylim(-.4,1.05)
```

```python
plot_pacf(arma_ts, method='ywm')
```
